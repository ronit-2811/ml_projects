```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sw3.csv",sep=';')
df.head()
```

```python
df["diff"]=((df["ClsPric"]-df["OpnPric"])/df["OpnPric"])*100
df
```

```python
loose = df.sort_values(by = "diff").head()
loose
```

```python
gain = df.sort_values(by = "diff",ascending=False).head()
gain
```

```python
vol = df.sort_values(by = "TtlTrfVal",ascending=False).head()
vol
```

```python
plt.bar(loose["TckrSymb"],loose["diff"])
plt.title('top 5 loosers')
plt.show()
```

```python
plt.bar(gain["TckrSymb"],gain["diff"])
plt.title('top 5 gainers')
plt.show()
```

```python
plt.bar(vol["TckrSymb"],vol["diff"])
plt.title('top 5 vol stocks')
plt.show()
```
