```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("task1.csv")
df
```

```python
df["X"].mean()
```

```python
X = df['X']
Y = df['Y']
x_mean = np.mean(X)
y_mean = np.mean(Y)
num = np.sum((X-x_mean) * (Y-y_mean))
den = np.sum((X-x_mean) ** 2)
m = num/den
c = y_mean - (m * x_mean)
y_p = (m * X) + c
print(y_p)
```

```python
ss_mean = np.sum((Y-y_mean)  ** 2)
ss_fit = np.sum((Y-y_p) ** 2)
rs = 1 - (ss_fit/ss_mean)
print(rs)
```

```python
print(f'slope = {m}')
print(f'intercept = {c}')
print(f'the best fit line is y = {m}x + {c}')
print(f'R^2 = {rs}')
```

```python
plt.plot(X,y_p)
plt.plot(X,Y,'o') 
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend(['best fit line','data points'])
plt.text(0, 0, f'R^2= {rs}', fontsize = 12, 
         bbox = dict(facecolor = 'white', alpha = 0.5))
plt.show()
```
